**Regularization**

this concept is used to reduce the overfitting Problem

- `They are 2 types of Regularizations `

`Ridge Regression [l2]`

`Lasso Regression [l1]`

```python
import numpy as np
import pandas as pd
import sklearn
```

```python
df=pd.read_csv('kc_house_data.csv')
df.head(7)
```

```python
df.shape
```

```python
df=df.drop(['id','date'],axis=1)
df.head(5)
```

```python
X=df.iloc[: ,1:]
y=df.iloc[ : ,0]
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=4)
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(X_train,y_train)
```

```python
reg.coef_
```

```python
print(f'Model Train Accuracy : {reg.score(X_train,y_train)}')
print(f'Model test Accuracy : {reg.score(X_test,y_test)}')
f=pd.DataFrame()
f['indepdent Colums']=X_train.columns
f['Model m_value ']=reg.coef_
f
```

**Ridge Regression**

```python
from sklearn.linear_model import Ridge
reg_ridge=Ridge()
reg_ridge.fit(X_train,y_train)
```

```python
print(f'Model Train Accuracy : {reg.score(X_train,y_train)}')
print(f'Model test Accuracy : {reg.score(X_test,y_test)}')

print('-------------------------------')

print(f'Ridge model Train Accuracy : {reg_ridge.score(X_train,y_train)}')
print(f'Ridge model Test Accuracy : {reg_ridge.score(X_test,y_test)}')

f=pd.DataFrame()
f['indepdent Colums']=X_train.columns
f['Model m_value ']=reg.coef_
f['Ridge model m_values']=reg_ridge.coef_
f
```

**Lasso Regression**

```python
from sklearn.linear_model import Lasso
reg_lasso=Lasso()
reg_lasso.fit(X_train,y_train)
```

```python
print(f'Model Train Accuracy : {reg.score(X_train,y_train)}')
print(f'Model test Accuracy : {reg.score(X_test,y_test)}')

print('-------------------------------')

print(f'Ridge model Train Accuracy : {reg_ridge.score(X_train,y_train)}')
print(f'Ridge model Test Accuracy : {reg_ridge.score(X_test,y_test)}')

print('-------------------------------')

print(f'Lasso model Train Accuracy : {reg_lasso.score(X_train,y_train)}')
print(f'Lasso model Test Accuracy : {reg_lasso.score(X_test,y_test)}')



f=pd.DataFrame()
f['indepdent Colums']=X_train.columns
f['Model m_value ']=reg.coef_
f['Ridge model m_values']=reg_ridge.coef_
f['Lasso model m_values']=reg_lasso.coef_
f
```
